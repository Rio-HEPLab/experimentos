```python
def Gaus(x, A, mean, sigma):
    return A * np.exp( -0.5*( (x - mean)/sigma )**2 )

def DoubleGaus(x, A, alpha, mean1, sigma1, mean2, sigma2):
    val_ = A * (          alpha * np.exp( -0.5*( (x - mean1)/sigma1 ) ** 2 ) +
                 ( 1. - alpha ) * np.exp( -0.5*( (x - mean2)/sigma2 ) ** 2 ) )
    return val_
```

```python
from scipy.optimize import curve_fit

# def plot_data( df, bins=100, hist_range=(-6000., 6000.), p0=(100.,0.,100.), fit_range=None ):
def plot_data( df, double_gaus = True, bins=100, hist_range=(-6000., 6000.), p0=(100., 0.5, 0., 100., 0., 100.), fit_range=None, label=None ):
    
    data = df['TDC_Diff']
    
    counts, bin_edges = np.histogram( data, bins=bins, range=hist_range )
    counts, bin_edges
    
    bin_centers = (bin_edges + 0.5)[:-1]
    
    sigma_counts = np.sqrt( counts )

    fig, axes = plt.subplots( 1, 2, figsize=(24,10) )
    axes[0].errorbar( bin_centers, counts, yerr=sigma_counts, fmt='ko', label=label )
    axes[0].set_xlabel( r'$\Delta t\,(\rm{ps})$', fontsize=18 )
    axes[0].set_ylabel( "NÃºmero de eventos", fontsize=18 )
    if label: axes[0].legend( loc='best', fontsize=16 )

    bin_centers_range = bin_centers
    counts_range = counts
    sigma_counts_range = sigma_counts
    if fit_range is not None:
        i_min = np.argmax( bin_centers >= fit_range[0] )
        i_max = np.argmax( bin_centers >= fit_range[1] ) + 1
        bin_centers_range = bin_centers[i_min:i_max]
        counts_range = counts[i_min:i_max]
        sigma_counts_range = sigma_counts[i_min:i_max]

#     popt, pcov = curve_fit(Gaus, bin_centers_range, counts_range, sigma=sigma_counts_range, p0=p0)
    popt, pcov = None, None
    if double_gaus:
        popt, pcov = curve_fit(DoubleGaus, bin_centers_range, counts_range, sigma=sigma_counts_range, p0=p0)
    else:
        popt, pcov = curve_fit(Gaus, bin_centers_range, counts_range, sigma=sigma_counts_range, p0=p0)
        
    print ( bin_centers_range, counts_range, sigma_counts_range )
    print ( popt, pcov )
    
    axes[1].errorbar( bin_centers, counts, yerr=sigma_counts, fmt='ko', label=label )
    axes[1].errorbar( bin_centers_range, counts_range, yerr=sigma_counts_range, fmt='bo' )
    axes[1].set_xlabel( r'$\Delta t\,(\rm{ps})$', fontsize=18 )
    axes[1].set_ylabel( "NÃºmero de eventos", fontsize=18 )
    if label: axes[1].legend( loc='best', fontsize=16 )
    
    func_range = hist_range
    if fit_range is not None: func_range = fit_range
        
    x_plot = np.linspace( *func_range, 200 )
#     axes[1].plot( x_plot, Gaus( x_plot, A=popt[0], mean=popt[1], sigma=popt[2] ), 'r' )
    if double_gaus:
        axes[1].plot( x_plot, DoubleGaus( x_plot, A=popt[0], alpha=popt[1], mean1=popt[2], sigma1=popt[3], mean2=popt[4], sigma2=popt[5] ), 'r' )
        axes[1].plot( x_plot, Gaus( x_plot, A=popt[0]*popt[1], mean=popt[2], sigma=popt[3] ), '--r' )
        axes[1].plot( x_plot, Gaus( x_plot, A=popt[0]*(1. - popt[1]), mean=popt[4], sigma=popt[5] ), '--b' )
    else:
        axes[1].plot( x_plot, Gaus( x_plot, A=popt[0], mean=popt[1], sigma=popt[2] ), 'r' )
        
    return (popt, pcov)
```

```python
def pol1(x, p0, p1):
   return p0 + p1*x
```

```python
m_idx_par = { -1: 1, 0: 2, 1: 4 }
idx_par = {}
for key_ in idx_gaus: idx_par[ key_ ] = m_idx_par[ idx_gaus[ key_ ] ]
print( idx_gaus )
print( idx_par )

x_values = np.array( list( distances.values() ) )
y_values = np.array( list( map( lambda key_: popt[ key_ ][ idx_par[ key_ ] ], popt ) ) ) # ps
y_values = y_values * 1e-03 # ns
sigma_y = np.sqrt( np.array( list( map( lambda key_: pcov[ key_ ][ idx_par[ key_ ],  idx_par[ key_ ]], pcov ) ) ) )
sigma_y = sigma_y * 1e-03
print ( x_values, y_values, sigma_y )

popt_lin, pcov_lin = curve_fit(pol1, x_values, y_values, sigma=sigma_y, p0=(0.,0.))
print ( popt_lin, pcov_lin )
```

```python
val = popt_lin[1]
err = np.sqrt( pcov_lin[1,1] )

velocity = 1. / val
err_velocity = ( ( ( 1. / val ) ** 2 ) * err )

s_ = r"v = {} \pm {}\,\rm{{cm}}/\rm{{ns}}".format( np.round( velocity, 2 ), np.round( err_velocity, 2 ) ) 
s_ = r"${}$".format( s_ )
print( "velocity = {:.2f} Â± {:.2f} cm/ns".format( np.round( velocity, 2 ), np.round( err_velocity, 2 ) ) )

fig = plt.figure( figsize=(12,10) )

plt.errorbar( x_values, y_values, yerr=sigma_y, fmt='ko' )
plt.plot( x_values, popt_lin[0] + popt_lin[1] * x_values, '-r')
plt.xlabel( "d (cm)", fontsize=18 )
plt.ylabel( r"$\overline{\Delta t}\,(\rm{ns})$", fontsize=18 )

plt.text( 0.1, 0.9,
          r"$v_{{\mu}} = {:.2f} \pm {:.2f}\,\rm{{cm}}/\rm{{ns}}$".format( np.round( velocity, 2 ), np.round( err_velocity, 2 ) ),
          fontsize=16, transform=plt.gca().transAxes )
```
