```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_0 = pd.read_table("dados-TDC/Teste-Dist-0.txt", header=None, names=("TDC1","TDC2"), sep='\s+')
df_1 = pd.read_table("dados-TDC/Teste-Dist-120.txt", header=None, names=("TDC1","TDC2"), sep='\s+')

df_0['TDC_Diff'] = df_0['TDC2'] - df_0['TDC1']
df_1['TDC_Diff'] = df_1['TDC2'] - df_1['TDC1']

print ( np.mean(df_0['TDC_Diff']), np.mean(df_1['TDC_Diff']) )

df = df_1
df.head(20)
```

```python
vel = (120. - 6.5) / ( np.mean(df_1['TDC_Diff']) - np.mean(df_0['TDC_Diff']) )

import scipy.constants
c_cm_ns = scipy.constants.c * (1.e+02 / 1.e+09 )

beta = vel / c_cm_ns

gamma = 1. / np.sqrt( 1. - beta**2 )

print (vel, beta, gamma)

mu_mass_MeV_c2 = 105.6583745

E_ave_MeV = mu_mass_MeV_c2 * gamma

print ( E_ave_MeV )
```

```python
data = df['TDC_Diff']
counts, bin_edges = np.histogram( data, bins=np.arange(-5,16) )
counts, bin_edges
```

```python
bin_centers = (bin_edges + 0.5)[:-1]
bin_centers
```

```python
sigma_counts = np.sqrt( counts )
sigma_counts
```

```python
def Gaus(x, A, mean, sigma):
   return A * np.exp( -0.5*( (x - mean)/sigma )**2 )
```

```python
from scipy.optimize import curve_fit

#i_min = 5
#i_max = 10

i_min = 8
i_max = 15

bin_centers_range = bin_centers[i_min:i_max]
counts_range = counts[i_min:i_max]
sigma_counts_range = sigma_counts[i_min:i_max]

popt, pcov = curve_fit(Gaus, bin_centers_range, counts_range, sigma=sigma_counts_range, p0=(1000.,0.,1.))
popt, pcov
```

```python
fig = plt.figure( figsize=(12,10) )

plt.plot( bin_centers_range, counts_range, 'ko' )

x_plot = np.linspace( -5., 15.0, 200 )
plt.plot( x_plot, Gaus( x_plot, A=popt[0], mean=popt[1], sigma=popt[2] ), 'r' )
```

```python
np.sqrt( pcov[1,1] ) 
```

```python
def plot_data(fileName, i_min=0, i_max=-1):
    
    df = pd.read_table(fileName, header=None, names=("TDC1","TDC2"), sep='\s+')
    df['TDC_Diff'] = df['TDC2'] - df['TDC1']
    
    data = df['TDC_Diff']
    counts, bin_edges = np.histogram( data, bins=np.arange(-5,16) )
    counts, bin_edges
    
    bin_centers = (bin_edges + 0.5)[:-1]
    
    sigma_counts = np.sqrt( counts )

    fig, axes = plt.subplots(1,2, figsize=(24,10))
    axes[0].plot( bin_centers, counts, 'ko' )
    
    bin_centers_range = bin_centers[i_min:i_max]
    counts_range = counts[i_min:i_max]
    sigma_counts_range = sigma_counts[i_min:i_max]

    popt, pcov = curve_fit(Gaus, bin_centers_range, counts_range, sigma=sigma_counts_range, p0=(1000.,0.,1.))
    
    print ( bin_centers_range, counts_range, sigma_counts_range )
    print ( popt, pcov )
    
    axes[1].plot( bin_centers_range, counts_range, 'ko' )

    x_plot = np.linspace( -5., 15.0, 200 )
    axes[1].plot( x_plot, Gaus( x_plot, A=popt[0], mean=popt[1], sigma=popt[2] ), 'r' )
    
    return (popt, pcov)
    
```

```python
popt_0, pcov_0 = plot_data("dados-TDC/Teste-Dist-0.txt", 5, 10)
```

```python
popt_120, pcov_120 = plot_data("dados-TDC/Teste-Dist-120.txt", 9, 15)
```

```python
popt_180, pcov_180 = plot_data("dados-TDC/Teste-Dist-180.txt", 11, 17)
```

```python
print ( popt_0 )
print ( pcov_0 )

print ( popt_120 )
print ( pcov_120 )

print ( popt_180 )
print ( pcov_180 )
```

```python
#vel = (120. - 6.5) / ( 6.66798593 - 2.32885495 )
dist_0 = 6.5
dist_120 = 120.

vel = (dist_120 - dist_0) / ( popt_120[1] - popt_0[1] )

import scipy.constants
c_cm_ns = scipy.constants.c * (1.e+02 / 1.e+09 )

beta = vel / c_cm_ns

gamma = 1. / np.sqrt( 1. - beta**2 )

print (vel, beta, gamma)

mu_mass_MeV_c2 = 105.6583745

E_ave_MeV = mu_mass_MeV_c2 * gamma

print ( E_ave_MeV )
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.array( (6.5, 120., 180.) ).reshape(-1,1)
train_y = np.array( (popt_0[1], popt_120[1], popt_180[1]) )
regr.fit(train_x, train_y)
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.figure(figsize=(12,10))
plt.plot(train_x, train_y, 'ko')
plt.plot(train_x, regr.coef_[0]*train_x + regr.intercept_, '-r')
```

```python
def pol1(x, p0, p1):
   return p0 + p1*x
```

```python
sigma_y = ( np.sqrt(pcov_0[1,1]), np.sqrt(pcov_120[1,1]), np.sqrt(pcov_180[1,1]) )  

popt, pcov = curve_fit(pol1, np.array( (6.5, 120., 180.) ), train_y, sigma=sigma_y, p0=(0.,0.))
print ( popt, pcov )
```
